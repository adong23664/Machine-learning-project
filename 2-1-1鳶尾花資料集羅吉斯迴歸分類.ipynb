
## 羅吉斯迴歸三類別分類器

羅吉斯迴歸（Logistic regression）是一種對數機率模型（Logit model）是離散選擇法模型之一，是用於分類而非回歸的線性模型。在此模型中，使用邏輯函數對描述單個試驗可能結果的概率進行建模。

下面顯示的是邏輯回歸分類器的決策邊界，該邊界位於鳶尾花資料及前兩個維度（花萼長度和寬度），資料點根據其標籤著色。


```python
#導入需要用到的套件
%matplotlib inline
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression  #導入邏輯斯回歸
from sklearn import datasets
```

```python
# 載入鳶尾花資料集
iris = datasets.load_iris()
X = iris.data[:, :2]  # 只用前兩個特徵值
Y = iris.target  #品種
```

```python
# 建立與訓練羅吉斯迴歸分類器
logreg = LogisticRegression(C=1e5)
logreg.fit(X, Y)
```

```python
# 畫出決策邊界(decision boundary)，同時也幫網格點 [x_min, x_max]x[y_min, y_max]上色
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# 幫結果上色
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# 幫訓練點上色
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()

#可以看出"邏輯斯回歸" 有分類錯誤現象
```

```python
# 載入鳶尾花資料集
iris = datasets.load_iris()
X = iris.data[:, 2:]  # 只用後兩個特徵值
Y = iris.target
# 建立與訓練羅吉斯迴歸分類器
logreg = LogisticRegression(C=100000)
logreg.fit(X, Y)
# 畫出決策邊界(decision boundary)，同時也幫網格點 [x_min, x_max]x[y_min, y_max]上色
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# 幫結果上色
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# 幫訓練點上色
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Petal length')
plt.ylabel('Petal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()

#可以看出"邏輯斯回歸" 有分類錯誤現象 但有改善
```
